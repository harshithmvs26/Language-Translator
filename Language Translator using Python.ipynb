```python
from tkinter import *
from tkinter import ttk
from googletrans import Translator, LANGUAGES

import tkinter as tk

def draw_gradient(canvas, width, height, start_color, end_color):
    r1, g1, b1 = canvas.winfo_rgb(start_color)
    r2, g2, b2 = canvas.winfo_rgb(end_color)
    r_ratio = (r2 - r1) / height
    g_ratio = (g2 - g1) / height
    b_ratio = (b2 - b1) / height

    for i in range(height):
        nr = int(r1 + (r_ratio * i))
        ng = int(g1 + (g_ratio * i))
        nb = int(b1 + (b_ratio * i))
        hex_color = f'#{nr//256:02x}{ng//256:02x}{nb//256:02x}'
        canvas.create_line(0, i, width, i, fill=hex_color)

root = tk.Tk()
root.geometry('1175x470')
root.resizable(0, 0)
root.title('üåê Language Translator by Simplilearn')

# Canvas for Gradient Background
canvas = Canvas(root, width=1175, height=470, highlightthickness=0)
canvas.pack(fill="both", expand=True)
draw_gradient(canvas, 1175, 470, '#1e3c72', '#2a5298')  # Blue gradient

# Main Heading
canvas.create_text(587, 35, text="üåê Language Translator", font=("Segoe UI", 28, "bold"), fill="white")

# Labels
label_input = Label(root, text="üìù Enter Text", font='Segoe 13 bold', bg='white', fg='black')
label_input.place(x=170, y=90)

Input_text = Entry(root, font='Segoe 12', width=60, bd=2, relief=RIDGE)
Input_text.place(x=30, y=130)

label_output = Label(root, text="üó£Ô∏è Output", font='Segoe 13 bold', bg='white', fg='black')
label_output.place(x=800, y=90)

Output_text = Text(root, font='Segoe 11', height=11, wrap=WORD, padx=5, pady=5, width=50, bd=2, relief=RIDGE)
Output_text.place(x=640, y=140)

# Language Dropdown
language = list(LANGUAGES.values())
dest_lang = ttk.Combobox(root, values=language, width=22, font='Segoe 11')
dest_lang.place(x=140, y=180)
dest_lang.set("üåç Choose Language")

# Translate Function
def Translate():
    translator = Translator()
    translated = translator.translate(text=Input_text.get(), dest=dest_lang.get())
    Output_text.delete(1.0, END)
    Output_text.insert(END, "üí¨ " + translated.text)

# Translate Button
style = ttk.Style()
style.configure('TButton', font=('Segoe UI', 12), padding=6)
trans_btn = Button(root, text='üîÑ Translate', font='Segoe 12 bold', pady=6, padx=12, command=Translate, bg='#ffcc00', fg='black', activebackground='#33cc33', relief=RAISED, borderwidth=3)
trans_btn.place(x=460, y=190)

# Footer Note
canvas.create_text(587, 450, text="üöÄ Powered by Google Translate API | UI by Simplilearn", font=("Segoe UI", 10), fill="white")

root.mainloop()
```
